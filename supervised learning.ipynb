```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
boston.keys()
```

```python
data = boston['data']
```

```python
feature_names = boston['feature_names']
feature_names
```

```python
X = pd.DataFrame(data, columns=feature_names)
X.head()
```

```python
X.info()
```

```python
target = boston['target']
```

```python
Y = pd.DataFrame(target, columns=['price'])
Y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, Y_train)
```

```python
Y_pred = lr.predict(X_test)
```

```python
check_test = pd.DataFrame({
    'y_test' : Y_test['price'],
    'y_pred' : Y_pred.flatten()
})
check_test.head(10)
```

```python
from sklearn.metrics import r2_score
```

```python
R2 = r2_score(Y_test['price'], Y_pred)
R2
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(n_estimators = 1000, max_depth = 12, random_state = 42)
```

```python
model.fit(X_train, Y_train.values[:, 0])
```

```python
Y_pred = model.predict(X_test)
```

```python
check_test = pd.DataFrame({
    'y_test' : Y_test['price'],
    'y_pred' : Y_pred.flatten()
})
check_test.head(10)
```

```python
R2 = r2_score(Y_test['price'], Y_pred)
R2
```

Модель дерева решений сработала лучше, т.к. r2 ближе к 1.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
```

```python
df.head(10)
```

```python
X = df[df.columns[:-1]]
X.head(10)
```

```python
y = pd.Series(df['Class'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100, stratify=y)
```

```python
X_train.info()
```

```python
parameters = [{'n_estimators': [10, 15],
              'max_features': np.arange(3, 5),
              'max_depth': np.arange(4, 7)}]
```

```python
clf = GridSearchCV(
estimator=RandomForestClassifier(random_state=100),
param_grid=parameters,
scoring='roc_auc',
cv=3
)
```

```python
clf.fit(X_train, y_train)
```

```python
clf.best_params_
```

```python
y_pred_proba = clf.predict_proba(X_test)
y_pred_proba = y_pred_proba[:, 1]
```

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(y_test, y_pred_proba)
```

```python
import matplotlib
import matplotlib.image as img
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
matplotlib.rcParams.update({'font.size': 14})
```

```python
plt.figure(figsize = (15,10))
sns.set(font_scale=1.4)
X_corr = X.corr()
X_corr = np.round(X_corr, 2)
X_corr[np.abs(X_corr) < 0.5] = 0
sns.heatmap(X_corr, annot=True, linewidths=.3, cmap='coolwarm')

plt.title('Correlation matrix')
plt.show()
```

```python
high_corr = X_corr[np.abs(X_corr) > 0.5]
```

```python
high_corr[np.abs(high_corr['target']) > 0.5]
```

```python
tar = high_corr['target']
```

```python
tar2 = tar[(np.abs(tar) > 0.5)][:-1]
tar2
```

```python
X.drop('target', axis = 1, inplace = True)
```
